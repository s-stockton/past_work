```
df.head(5)
```

```
tableids = ['B01001', 'B19001', 'B03003']
```

```
tableidstring = ','.join(tableids)
```

```
requesturl = 'https://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=310|01000US' % tableidstring
```

```
loadedjson = requests.get(requesturl)
loadedjson
type(loadedjson)
```

```
parsedjson = loadedjson.json()
```

```
#parsedjson
```

```
hisp = pd.DataFrame({ 'hisp': { id_: data['B03003']['estimate']['B03003003'] for id_, data in parsedjson['data'].items() }})
```

```
hisp.sort_values(by='hisp', ascending = False)
```

```
hisp['hisp_percent'] = hisp['hisp']/5973798
```

```
hisp.sort_values(by='hisp_percent', ascending = False)
```

```
m1819 = pd.DataFrame({ 'm1819': { id_: data['B01001']['estimate']['B01001007'] for id_, data in parsedjson['data'].items() }})
m20 = pd.DataFrame({ 'm20': { id_: data['B01001']['estimate']['B01001008'] for id_, data in parsedjson['data'].items() }})
m21 = pd.DataFrame({ 'm21': { id_: data['B01001']['estimate']['B01001009'] for id_, data in parsedjson['data'].items() }})
m2224 = pd.DataFrame({ 'm2224': { id_: data['B01001']['estimate']['B01001010'] for id_, data in parsedjson['data'].items() }})
m2529 = pd.DataFrame({ 'm2529': { id_: data['B01001']['estimate']['B01001011'] for id_, data in parsedjson['data'].items() }})
m3034 = pd.DataFrame({ 'm3034': { id_: data['B01001']['estimate']['B01001012'] for id_, data in parsedjson['data'].items() }})
f1819 = pd.DataFrame({ 'f1819': { id_: data['B01001']['estimate']['B01001031'] for id_, data in parsedjson['data'].items() }})
f20 = pd.DataFrame({ 'f20': { id_: data['B01001']['estimate']['B01001032'] for id_, data in parsedjson['data'].items() }})
f21 = pd.DataFrame({ 'f21': { id_: data['B01001']['estimate']['B01001033'] for id_, data in parsedjson['data'].items() }})
f2224 = pd.DataFrame({ 'f2224': { id_: data['B01001']['estimate']['B01001034'] for id_, data in parsedjson['data'].items() }})
f2529 = pd.DataFrame({ 'f2529': { id_: data['B01001']['estimate']['B01001035'] for id_, data in parsedjson['data'].items() }})
f3034 = pd.DataFrame({ 'f3034': { id_: data['B01001']['estimate']['B01001036'] for id_, data in parsedjson['data'].items() }})
```

```
ages1834 = pd.concat([m1819, m20, m21, m2224, m2529, m3034, f1819, f20, f21, f2224, f2529, f3034], axis=1) 
```

```
ages1834
```

```
ages1834['total'] = ages1834.sum(axis = 1, skipna = True) 
```

```
ages1834
```

```
ages1834.sort_values(by='total', ascending = False)
```

```
ages1834['ages_percent'] = ages1834['total']/4797805
```

```
ages1834.sort_values(by='ages_percent', ascending = False)
```

```
hh50k = pd.DataFrame({ '50k': { id_: data['B19001']['estimate']['B19001011'] for id_, data in parsedjson['data'].items() }})
```

```
hh50k
```

```
hh5075k = pd.DataFrame({ '50-75k': { id_: data['B19001']['estimate']['B19001012'] for id_, data in parsedjson['data'].items() }})
```

```
hh5075k
```

```
sumhhinc = hh50k.add(hh5075k, fill_value=0)
```

```
sumhhinc['total'] = sumhhinc['50k']+sumhhinc['50-75k']
```

```
sumhhinc.sort_values(by='total', ascending = False)
```

```
sumhhinc['hhinc_percent'] = sumhhinc['total']/1050888
```

```
sumhhinc.sort_values(by='hhinc_percent', ascending = False)
```

```
highest_percent = pd.concat([hisp['hisp_percent'], ages1834['ages_percent'], sumhhinc['hhinc_percent']], axis=1, keys=['hisp_percent', 'ages_percent', 'hhinc_percent'])
```

```
highest_percent
```

```
highest_percent['total'] = highest_percent['hisp_percent']+highest_percent['ages_percent']+highest_percent['hhinc_percent']
```
